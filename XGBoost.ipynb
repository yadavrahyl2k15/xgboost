```python
import pandas as pd
import numpy as np
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("/content/diabetes.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
x=data.drop(columns="Outcome")
y=data["Outcome"]
```

```python
scaler=StandardScaler()
scaled_x=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled_x,y,test_size=0.33,random_state=0)
```

```python
model=XGBClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
param_girid={
    "learning_rate":[1,0.5,0.1,0.01,0.001],
    "max_depth":[3,5,10,20],
    "n_estimators":[10,90,100,110]
}
```

```python
girid=GridSearchCV(XGBClassifier(),param_girid,verbose=3)
```

```python
girid.fit(x_train,y_train)
```

```python
girid.best_params_
```

```python
model_1=XGBClassifier(learning_rate=0.01,max_depth=3,n_estimators=110)
model_1.fit(x_train,y_train)
```

```python
y_predict_1=model_1.predict(x_test)
```

```python
accuracy_1=accuracy_score(y_test,y_predict_1)
accuracy_1
```
